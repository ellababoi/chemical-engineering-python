Model the concentration of reactant A as it undergoes a first order reaction in a stirred tank that has two input points.

```python
import numpy
import pylab
from scipy.integrate import odeint

# model kinetic parameters
V = 50.0 # mol/(L s)
k = 0.35/60 # min-1
Q1 = 10.0
Ca1 = 2.0
Q2 = 8
Ca2 = 5

# ODE definition
def df(c,t):
    Ca = c
    dCadt = ((Ca1*Q1 + Ca2*Q2 - (Q1+Q2)*Ca)/V)-k*Ca
    return dCadt
        
# initial condition
c0 = 0.73
t = numpy.linspace(0,15.0,16)
sol = odeint(df, c0, t)
pylab.plot(t,sol)
pylab.xlabel("time")
pylab.ylabel("concentration")
```
